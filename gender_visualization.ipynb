### Gender Visualization Notebook

Import the necessary modules:

```python
import pandas as pd
import plotnine as p9
```

Load in the reference CSV as a dataframe and inspect it:

```python
df = pd.read_csv("articles_gender_validated.csv")
df.head()
```

```python
gender_by_year = df.groupby(['year', 'auth_gender']).size().to_frame().reset_index()
gender_by_year.columns = ['year', 'gender', 'count']
gender_by_year.head()
```

Number of Authors by Gender and Year (All Articles)

```python
gby_all = (p9.ggplot(data=gender_by_year,
           mapping=p9.aes(x='year', y='count', group='gender'))
    + p9.geom_point(mapping=p9.aes(color='gender'))
    + p9.geom_line(mapping=p9.aes(color='gender'))
    + p9.labs(title="Number of Authors by Gender and Year (All Articles)",
           color="Gender", x="Year", y="Number of Authors")
    + p9.theme_classic()
    )
gby_all.save("gender_by_year_all.png", format='png')
gby_all
```

Gender Ratio of Authors by Year (All Articles)

```python
grby_all = (p9.ggplot(data=gender_by_year, mapping=p9.aes(x='year', y='count', fill='gender'))
           + p9.geom_bar(position="fill", stat="identity")
           + p9.labs(title="Gender Proportion of Authors by Year (All Articles)",
                     fill="Gender", x="Year", y="Gender Proportion")
           + p9.theme_classic()
           )
grby_all.save("gender_proportion_by_year_all.png", format='png')
grby_all
```
